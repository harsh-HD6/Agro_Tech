```python
def get_ctr(f):
    ctr = np.genfromtxt(img2txt_name(f))
    print(ctr)
    return ctr[0]
```

```python
def img2txt_name(f): return path/f'{str(f)[:-4]}txt'
```

```python
data = (ImageList.from_folder(path)
        .split_by_rand_pct()
        .label_from_func(get_ctr)
        .transform(get_transforms())
        .databunch().normalize(imagenet_stats)
       )
```
